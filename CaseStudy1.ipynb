# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
Sales_gender=car['Buyer Gender'].value_counts().to_frame().reset_index() # Plotting the bar graph with x and y coordinate values
plt.bar(Sales_gender['index'],Sales_gender['Buyer Gender'],color= ['orange','green']) # Providing different properties
plt.title('Gender Count') # title for bargraph
plt.xlabel('Gender') # naming x coordinate
plt.ylabel('Count') # naming y Coordinate
```

```python
Sales_gender # no of males and females
```

#### The insight from the bar graph is that both the males and females participated in the sales are almost same and only having slight difference

# 4. Find the top 5 cars based on their sales price.
